```python
# Imports.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read csv file.
olympics_df = pd.read_csv('athlete_events.csv')
```

```python
usa_gold = olympics_df[(olympics_df['NOC'] == 'USA') & (olympics_df['Medal'] == 'Gold')]
usa_gold_count = usa_gold['Medal'].count()
usa_silver = olympics_df[(olympics_df['NOC'] == 'USA') & (olympics_df['Medal'] == 'Silver')]
usa_silver_count = usa_silver['Medal'].count()
usa_bronze = olympics_df[(olympics_df['NOC'] == 'USA') & (olympics_df['Medal'] == 'Bronze')]
usa_bronze_count = usa_bronze['Medal'].count()
x = np.array([usa_gold_count, usa_silver_count, usa_bronze_count])
my_labels = ['Gold', 'Silver', 'Bronze']
my_colors = ['Gold', 'Silver', 'DarkGoldenRod']

def format_label(pct, data):
    count = int(pct / 100. * sum(data))
    return f'{count} \n{pct:.1f}%'

plt.pie(x, autopct = lambda pct: format_label(pct, x), startangle=90, labels = my_labels, colors = my_colors)

plt.title("Breakdown of USA Olympic Medals Won")

plt.show()
```

```python
country_list = olympics_df['NOC'].unique().tolist()

medals = ['Gold', 'Silver', 'Bronze']
won_medal = olympics_df[olympics_df['Medal'].isin(medals)]
country_list_won_medal = won_medal['NOC'].unique().tolist()

no_medal = []
for i in country_list:
    if i not in country_list_won_medal:
        no_medal.append(i)
print("There are", len(no_medal), "countries who have participated in at least one Olympic games and never won a single medal.\n")
print(no_medal)


won_gold = olympics_df[(olympics_df['Medal'] == 'Gold')]
country_list_won_gold = won_gold['NOC'].unique().tolist()

no_gold = []
for i in country_list:
    if i not in country_list_won_gold:
        no_gold.append(i)
print("\n\nThere are", len(no_gold), "countries who have participated in at least one Olympic games and never won a gold medal.\n")
print(no_gold)
```

```python
medals = ['Gold', 'Silver', 'Bronze']
age_medal_winners = olympics_df[olympics_df['Medal'].isin(medals)]

x = age_medal_winners[['Age']].value_counts().reset_index(name = 'count')

age = x['Age'].tolist()
medal_count = x['count'].tolist()

plt.bar(age, medal_count)
plt.xticks(np.arange(0, 70, 5))
plt.xlabel("Age", fontweight = 'bold')
plt.yticks(np.arange(0, 4000, 500))
plt.ylabel("Number of Medals Won", fontweight = 'bold')
plt.title(label = "Olympic Medals Won by Age", fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
age_medal_winners = olympics_df[olympics_df['Medal'] == 'Gold']

x = age_medal_winners[['Age']].value_counts().reset_index(name = 'count')

age = x['Age'].tolist()
medal_count = x['count'].tolist()

plt.bar(age, medal_count)
plt.xticks(np.arange(0, 70, 5))
plt.xlabel("Age", fontweight = 'bold')
plt.yticks(np.arange(0, 1200, 100))
plt.ylabel("Number of Gold Medals Won", fontweight = 'bold')
plt.title(label = "Olympic Gold Medals Won by Age", fontsize = 15, fontweight = 'bold')
plt.show()
```

```python
summer_participants = olympics_df[olympics_df['Season'] == 'Summer']

male_summer_participants = summer_participants[summer_participants['Sex'] == 'M']
x = male_summer_participants[['Year']].value_counts().reset_index(name = 'count')
x = x.sort_values(by = ['Year'])

female_summer_participants = summer_participants[summer_participants['Sex'] == 'F']
y = female_summer_participants[['Year']].value_counts().reset_index(name = 'count')
y = y.sort_values(by = ['Year'])

male_summer_year = x['Year'].tolist()
male_summer_count = x['count'].tolist()

female_summer_year = y['Year'].tolist()
female_summer_count = y['count'].tolist()


plt.plot(male_summer_year, male_summer_count, marker = 'o', linestyle = 'dashed')
plt.plot(female_summer_year, female_summer_count, marker = 'o', linestyle = 'dashed')
plt.xlabel("Year", fontweight = 'bold')
plt.yticks(np.arange(0, 9000, 1000))
plt.ylabel("Number of Athletes", fontweight = 'bold')
plt.title(label = "Olympic Summer Athletes by Sex", fontsize = 15, fontweight = 'bold')
plt.legend(["Male", "Female"])
plt.show()
```

```python
winter_participants = olympics_df[olympics_df['Season'] == 'Winter']

male_winter_participants = winter_participants[winter_participants['Sex'] == 'M']
x = male_winter_participants[['Year']].value_counts().reset_index(name = 'count')
x = x.sort_values(by = ['Year'])

female_winter_participants = winter_participants[winter_participants['Sex'] == 'F']
y = female_winter_participants[['Year']].value_counts().reset_index(name = 'count')
y = y.sort_values(by = ['Year'])

male_winter_year = x['Year'].tolist()
male_winter_count = x['count'].tolist()

female_winter_year = y['Year'].tolist()
female_winter_count = y['count'].tolist()


plt.plot(male_winter_year, male_winter_count, marker = 'o', linestyle = 'dashed')
plt.plot(female_winter_year, female_winter_count, marker = 'o', linestyle = 'dashed')
plt.xlabel("Year", fontweight = 'bold')
plt.yticks(np.arange(0, 3500, 500))
plt.ylabel("Number of Athletes", fontweight = 'bold')
plt.title(label = "Olympic Winter Athletes by Sex", fontsize = 15, fontweight = 'bold')
plt.legend(["Male", "Female"])
plt.show()
```
